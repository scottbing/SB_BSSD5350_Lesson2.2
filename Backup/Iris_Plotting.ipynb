```python
#find the index of the first virginiaca (lable 2)
# convert numpy array to python list
# and find first index of 2
cutoff = list(iris.target).index(2)
#copy onl the data before that
binary_irises = iris.data[:cutoff]
binary_targets = iris.target[:cutoff]
```

```python
x_index = 0 # only plotting the first measurement of each flower
y_index = 1 # along withthe second measurement of each flower. ignoring 3 and 4

plt.figure(figsize=(10, 5)) #width and height of graph
#walks through the data taking 0 for x and 1 element for y of each row
#c is color based on the target lable for current row index
plt.scatter(binary_irises[:, x_index], binary_irises[:, y_index], c=binary_targets)
plt.title('Iris Dataset Scatterplot')
plt.x_label(iris.feature_names[x_index])
plt.y_label(iris.feature_names[y_index])

plt.show() #plot willnot dipaly without show
```
